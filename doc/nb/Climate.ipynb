# Climate

```python
# imports
from matplotlib import pyplot as plt

import iris
from iris import quickplot
```

```python
%matplotlib notebook
```

# Load

```python
test_climate_file = '/home/xavier/Projects/Oceanography/MHWs/db/test_climate.nc'
```

```python
cube = iris.load(test_climate_file)
```

```python
cube
```

## Parse

```python
seasonal = cube[0]
seasonal
```

## Plot

```python
time = iris.Constraint(day=0)
```

```python
jan1 = seasonal.extract(time)
```

```python
jan1
```

```python

# Draw the contour with 25 levels.
quickplot.contourf(jan1, 25)

# Add coastlines to the map created by contourf.
plt.gca().coastlines()

plt.show()
```
