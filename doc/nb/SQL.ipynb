# SQL Fussing

```python
# imports
import numpy as np
import pandas
import sqlalchemy
```

# Full Table first

## Build it

```python
full_tbl = pandas.DataFrame.from_dict({'lat': np.linspace(-90., 90.), 'lon': np.linspace(0., 360.)})
```

```python
full_tbl[0:10]
```

## Do it

```python
engine = sqlalchemy.create_engine('sqlite:///foo.db')
```

```python
full_tbl.to_sql('coord', con=engine)
```

## Read it

```python
engine2 = sqlalchemy.create_engine('sqlite:///foo.db')
```

```python
full_tbl2 = pandas.read_sql_table('coord', con=engine2)
```

```python
full_tbl2[0:10]
```

```python
full_tbl.dtypes
```

```python
full_tbl2
```

# In steps

```python
engine3 = sqlalchemy.create_engine('sqlite:///foo3.db')
```

```python
sub_tbl1 = pandas.DataFrame.from_dict({'lat': np.linspace(-90., 0.), 'lon': np.linspace(0., 180.)})
```

```python
sub_tbl1.to_sql('coord', con=engine3, if_exists='append')
```

```python
sub_tbl2 = pandas.DataFrame.from_dict({'lat': np.linspace(0., 90.), 'lon': np.linspace(180., 360.)})
```

```python
sub_tbl2.to_sql('coord', con=engine3, if_exists='append')
```

## Read

```python
engine4 = sqlalchemy.create_engine('sqlite:///foo3.db')
```

```python
full_tbl3 = pandas.read_sql_table('coord', con=engine4)
```
